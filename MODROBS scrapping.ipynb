```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the website to search
url = 'https://www.cgc.edu.in/funded-projects'

# Send a request to the website and fetch the HTML content
response = requests.get(url)
html_content = response.text

# Save the HTML content to a file
with open('website.html', 'w', encoding='utf-8') as file:
    file.write(html_content)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Search for the keyword "MODROB" in the HTML content
keyword = 'MODROBS'
found = False
for tag in soup.find_all():
    if keyword in str(tag):
        print(f"Keyword '{keyword}' found in the following tag:")
        print(tag)
        print()
        found = True

# If the keyword is not found, print a message
if not found:
    print(f"No occurrence of '{keyword}' found on the website.")
```

```python
import requests

# Define the keyword to search for
keyword = "MODROBS"

# Define the URL to start the search from
start_url = "https://vidwan.inflibnet.ac.in/"

# Send a GET request to the start URL
response = requests.get(start_url)

# Check if the request was successful
if response.status_code == 200:
    # Extract the HTML content from the response
    html_content = response.text

    # Check if the keyword is present in the HTML content
    if keyword in html_content:
        print(f"Keyword '{keyword}' found in {start_url}")
    else:
        print(f"Keyword '{keyword}' not found in {start_url}")
else:
    print(f"Failed to retrieve content from {start_url}. Status code: {response.status_code}")
```

```python

# Import the beautifulsoup 
# and request libraries of python.
import requests
import bs4
  
# Make two strings with default google search URL
# 'https://google.com/search?q=' and
# our customized search keyword.
# Concatenate them
text= "MODROBS"
url = 'https://google.com/search?q=' + text
  
# Fetch the URL data using requests.get(url),
# store it in a variable, request_result.
request_result=requests.get( url )
  
# Creating soup from the fetched request
soup = bs4.BeautifulSoup(request_result.text,
                         "html.parser")
print(soup)
```

```python
# soup.find.all( h3 ) to grab 
# all major headings of our search result,
heading_object=soup.find_all( 'h3' )
  
# Iterate through the object 
# and print it as a string.
for info in heading_object:
    print(info.getText())
    print("------")
```

```python
from googlesearch import search

def scrape_google(query, num_results=10):
    # Create an empty list to store the extracted links
    links = []

    # Iterate through the search results using the googlesearch package
    for url in search(query, num_results=num_results):
        # Add the URL to the list of links
        links.append(url)

    # Return the list of links
    return links

# Example usage:
query = "MODROBS fund received"
num_results = 50
links = scrape_google(query, num_results)
print(links)
```

```python
from googlesearch import search
import time

def scrape_google(query, num_results=10):
    # Create an empty list to store the extracted links
    links = []

    # Iterate through the search results using the googlesearch package
    for url in search(query, num_results=num_results):
        # Add the URL to the list of links
        links.append(url)

        # Add a delay of 2 seconds between each request to avoid hitting rate limits
        time.sleep(2)

    # Return the list of links
    return links

# Example usage:
query = "MODROBS fund received"
num_results = 10
links = scrape_google(query, num_results)
print(links)
```
